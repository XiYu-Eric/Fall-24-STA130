Attributes vs Methods:

Attributes (like df.shape) provide information about the DataFrame without parentheses, while methods (like df.describe()) perform actions or computations and require parentheses.
df.describe():

Provides summary statistics such as count, mean, std, min, max, and percentiles (25%, 50%, 75%) for numerical columns.
df.describe() only shows statistics for numeric columns by default and may show fewer columns than df.shape suggests.

Use of df.groupby() and describe():
